# stack


* push(item) - add item to top - O(1)
* pop() - remove and return top item - O(1)
* top()/peek() - view top item without removing - O(1)
* empty() - check if stack is empty - O(1)
* size() - get number of elements - O(1)

```python
// C++ built-in stack
#include <iostream>
#include <stack>
using namespace std;


int main() {
   stack<int> st;
   st.push(10);
   st.push(20);
   st.push(15);
   st.push(1);

   cout << st.top() << endl; // 1
   st.pop();
   st.pop();
   cout << st.top() << endl; // 20
   cout << st.size() << endl;
   cout << st.empty();
   return 0;
}
```

```python
#include <iostream>
#include <stack>
using namespace std;


/*Check if brackets are properly balanced: 
(), [], {}
*/

bool is_ok(string s) {
    stack<char> st;

    for (int i = 0; i < s.size(); i++) {
        if (s[i] == '(' || s[i] == '{' || s[i] == '[')
            st.push(s[i]);
        else {
            if (st.empty())
                return false;
            if (s[i] == '}' && st.top() != '{')
                return false;
            if (s[i] == ']' && st.top() != '[')
                return false;
            if (s[i] == ')' && st.top() != '(')
                return false;
            st.pop();                
        }
    }
    return (st.empty());
}

int main() {
    string s;
    cin >> s;
    if (is_ok(s))
        cout << "YES";
    else
        cout << "NO";
    return 0;
}
```

# queue



Core Operations:
* FIFO (First In First Out) principle

* push()/enqueue(item) - add item to rear - O(1)
* pop()/dequeue() - remove front item - O(1)
* front() - view front item without removing - O(1)
* empty() - check if queue is empty - O(1)
* size() - get number of elements - O(1)

```python

#include <queue>
#include <iostream>
using namespace std;

void demo_stl_queue() {
    queue<int> q;
    
    // Add elements
    q.push(10);
    q.push(20);
    q.push(30);
    
    cout << "Queue size: " << q.size() << endl;
    cout << "Front element: " << q.front() << endl;
    
    // Remove elements
    while (!q.empty()) {
        cout << "Removing: " << q.front() << endl;
        q.pop();
    }
}
```

# deque


Concept: Combination of stack and queue - can add/remove from both ends efficiently.

Core Operations:

* push_back(item) - add item to back - O(1)
* push_front(item) - add item to front - O(1)
* pop_back() - remove back item - O(1)
* pop_front() - remove front item - O(1)

```python

deque<int> q;
q.push_back(10); 
q.push_back(20);
q.push_front(30);
q.push_front(40);
```
